```python
def remove_stopwords(text):
    words = text.split()  # Split the text into words
    stop_words = set(stopwords.words('english'))  # Get the set of English stop words
    filtered_words = [word for word in words if word.lower() not in stop_words]  # Remove stop words
    return ' '.join(filtered_words)  # Join the filtered words back into a string

tech = remove_stopwords(tech)
print(tech)
```

```python
def remove_special_characters(text):
    pattern = r'[^a-zA-Z\s]'
    text_without_special_chars = re.sub(pattern, '', text)
    return text_without_special_chars

file_content = remove_special_characters(tech)
print(tech)
```

```python
def count_word_frequencies(text):
    # Tokenize the text and convert it to lowercase
    words = re.findall(r'\w+', text.lower())
    word_counts = Counter(words)
    return word_counts

# Count word frequencies in the text
word_frequencies = count_word_frequencies(tech)

for word, count in word_frequencies.items():
    print(f"{word}: {count}")
```

```python
from textblob import TextBlob
nltk.download('brown')
nltk.download('averaged_perceptron_tagger')
```

```python
def extract_text_structure(text):
    # Create a TextBlob object from the input text
    blob = TextBlob(text)

    # Extract the sentences and their corresponding noun phrases
    clausal = []
    noun_phrase = []
    for sentence in blob.sentences:
        for np in sentence.noun_phrases:
            noun_phrase.append(np)
        tagged = nltk.pos_tag(sentence.words)
        subject = None
        for word, tag in tagged:
            if tag in ('NN', 'NNS') and (subject is None or tag=='NNS'):
                subject = word
        clausal.append((str(sentence), subject))

    # Count the frequency of each noun phrase
    noun_phrase_freq = Counter(noun_phrase)

    # Return the list of clauses and noun phrases with their frequencies
    return clausal, noun_phrase_freq
extract_text_structure(tech)
```
